```python
def mean_val(a) :  # 변수를 굳이 생성할 필요 X
    return sum(a) / len(a)
```

```python
def mean_val(a) : return sum(a) / len(a)  # 들여쓰기가 한 줄이면, 들여쓰기 필요 X
# def -> 함수 정의 부분
# mean_val -> 함수 이름 부분
# (a) -> 소괄호는 함수 이름과 함수에 들어가는 재료 구분, a는 함수의 재료
# : -> 왼쪽과 오른쪽 영역을 구분하는 기준
# return -> 함수 밖으로 나오는 값 정의 부분(코드가 여러 줄일 경우, 어떤 값을 어떤 조건에 반환할 것인지 중요, 현 코드에서는 중요 X)
# sum(a) / len(a) -> 함수 밖으로 나오는 값
```

```python
lambda a : sum(a) / len(a)  # def 가 아닌 lambda 식으로 표현한 평균 함수
# 콜론 왼쪽 것이 들어가서 오른쪽 것이 나온다.(lambda 도 함수!)
# 숫자뿐만 아니라 문자에서도 활용 가능, 무궁무진

# lambda 의 의의
# 1. 데이터 분석 시 유용하게 활용 가능 -> 머릿속 아이디어를 그때 그때 구현할 때 자주 활용(간단함)
# 2. 본인의 코딩 실력과 상관없이, 다른 사람의 코드를 볼 때, 타인이 자주 사용하는 문법
```

### 데이터 프레임(Data Frame)
#### table 이라고도 부름(R의 read.table 함수)
* 데이터를 가로 X 세로 직사각형 모양으로 틀을 갖춰 저장 or 다룹니다(정형 데이터, tabular data)
* 가로(행)를 row라 하고, 세로(열)를 column이라 합니다.
* 한 row 안의 데이터는 딕셔너리처럼 들어있습니다(한 key에 대한 value가 1개씩)
* 한 column 안의 데이터는 리스트처럼 들어있습니다.


### Pandas - 파이썬에서 데이터 프레임을 다루는 도구(유일무이)
##### PANel + DAta + analysiS
* 통계학에서, 데이터 프레임 형태의,
* 여러 개체들을 복수의 시간에 걸쳐서 추적하여 얻는 데이터 자료를 panel data(longitudinal data, 종단 자료)라고 함
* 추천 도서 : 파이썬 라이브러리를 활용한 데이터 분석(입문 단계 X, 머신러닝 수업 이후 참고서적으로 활용 가능)

##### row 관점으로 데이터 프레임 생성
* 빈 리스트 생성 후, 딕셔너리(row 측면)를 하나씩 추가함

```python
import pandas as pd  # pandas 도구를 불러와 pd 라는 축약어로 사용, 공식 문서와 다수의 사람들이 해당 축약어 사용
all_list = []  # 아무것도 없는 빈 리스트 생성 -> 이후 채워넣을 것
all_list.append({"data" : "2020-01-01", "dayofweek" : "wednesday"})  # 타 언어에서는 MAP 혹은 JSON이라 함
# 딕셔너리는 말 그대로 사전 형식이므로 key 값으로 value를 조회, 역순은 불가능
all_list.append({"data" : "2020-01-02", "dayofweek" : "thursday"})
all_list.append({"data" : "2020-01-03", "dayofweek" : "friday"})
all_list
```

```python
pd.DataFrame(all_list)
```

##### row 관점이 아닌, column 관점으로 데이터 프레임 생성
* 리스트(column) 여러 개 생성 후, key 를 설정하여 리스트 내용을 딕셔너리로 묶음

```python
dates = ["2020-01-01", "2020-01-02", "2020-01-03"]
yo_il = ['wednesday', 'thursday', 'friday']
data = {"date": dates, "dayofweek": yo_il}
data
```

```python
df = pd.DataFrame(data)
df
```

##### 불리언 인덱싱 : 불리언(참/거짓) 정보를 이용하여 조건을 만족하는(True) row 선택

```python
df.loc[df['holiday'] == 1]  # 조건 생성
# 전체 데이터 프레임 df 중에서 'holiday 컬럼 기준 == 1' 조건이 True 인 row만 선택.
```

```python
df['cold'] = 0  # 'cold' 컬럼을 생성할 때, 미리 초기값을 설정하여 NaN(빈 값)을 방지.
df.loc[df['temp'] < 1, 'cold'] = 1
# .loc[row 정보, column 정보]
# df['temp'] < 1 조건을 만족하는 row 를 따로 선택하고
# 해당 row 에 대하여 'cold' 컬럼에 1을 넣는다
df.head()
```
