### Business Understanding

**한 문장 요약 : 무엇을 하고 싶은가?** (나의 관점이 아닌 기획자 / 총책임자 관점)  
\--> label 을 명확히 정하는 것.
* 분야별(산업군)로 관심사(KPI, 목표)가 다르다.  
\-> 공장(제조산업)은 생산성 극대화(시간, 돈, 자원 투입 대비 생산량)  
\-> 쇼핑몰(리테일)은 재구매율 증가, 매출액 증대 등
* 과거 학술 / 기술적으로 선행연구가 무엇이 있었나 파악(업계의 고민 파악)
* **분석 방향성 및 목표 설정**(업계의 고민 해결에 도움이 되는 목표 / 방향인가?)

일반적으로 Business Understanding 은 팀 단위의 분석에서는 이미 결론이 난 경우가 많음. 혼자서 or 의뢰의 목적이 비지니스 로직인 경우, 유용함.  
그러나 **분석 방향성 및 목표 설정** 경험이 있어야 어느 정도 실력이 인정됨.  
\--> 따라서 인터넷에 있는 자료를 그대로 가져오지 말고 여러 개 합쳐서 나만의 데이터 셋을 만들어본다.

### Data Preparation
##### 데이터 준비 단계(데이터 탐색과 묶어서 생각하기도 함)
* 준비: 나중의 일(modeling)을 미리 대비. 모델링은 알고리즘에 데이터를 넣고 돌리는 단계.
* 알고리즘에 넣을 데이터를 '적절하게' 준비시킴.  
= 데이터를 분석 모형(알고리즘)에 넣기 위해 형태를 맞추는 것(알고리즘마다 모양 다르기 때문)
* 기계학습(Machine Learning) - 학습? 데이터에 들어있는 패턴 파악.
 * 학습(공부) 방식에 따라 크게 지도학습 / 비지도학습으로 나뉘어짐.
 * 지도학습이란? 채점이 가능한 상황. 문제와 정답이 존재.  
\-> 지도학습에서는 데이터를 문제와 정답 모음으로 봄.
 * 비지도학습이란? 채점이 불가능, 정답이 없음.  
\-> 쇼핑몰에서 추천 시스템 구축한 경우, 특정 물건을 추천할 때 정답이라서 추천 X  
\-> 고객군 그룹화 - 우리 쇼핑몰의 1만명 회원들을 5개의 그룹으로 묶는다?  
\--> 데이터의 패턴상 5개로 묶인 것이지, 5개가 정답이라서 묶은 것이 아님.
* 데이터 준비단계에서 중요하게 작업하는 것 - **train / test split**  
\--> 공부용과 시험용 데이터 셋 분리(지도학습)  
\--> 공부했던 문제가 시험문제와 중복되지 않게끔  
\--> 과적합(Overfit) 방지, 시험결과를 신뢰할 수 없기 때문

### Model Planning
##### 분석 모형 기획(알고리즘 후보군 선정)
* 같은 효과(기본적인 목적)를 내는 알고리즘이 다양합니다.
* 알고리즘 선택 시, 3가지 방향성을 고려합니다.  
\--> 속도 / 예측력 / 해석력
* 속도 - 비즈니스 프로세스(비즈니스 이해)에서 결정. 보통 예측력과 trade-off.  
보통 예측력은 높을수록 좋지만, 속도의 경우 일정 속도 이상을 요구하는 경우가 많아 무시할 수 없음.
* 예측력 - 알고리즘 결과가 실제 값과 얼마나 같은지? 정답을 얼마나 잘 맞추는지?
* 해석력 - Interpretability, Explainability - 결과 해석 용이성.  
\--> 해석력이 안 좋으면 블랙박스라 함. 좋으면 화이트박스.  
\--> 내부 계산과정이 복잡할수록 해석력이 떨어짐(자동적으로 속도 감소).
* 기술통계 분야 알고리즘의 경우, 속도와 해석력이 우수해야 함.
* XAI - (eXplainable AI) : 블랙박스 모델을 해석하는 알고리즘(왜 이런 결과를 냈나?)  
(최신반) PDP, ICE, Shapley value, Permutation Importance -> 변수 선택법 참고  
(수업 X) LIME, LOCO 등

일반적으로 예측력이 높으면 속도와 해석력이 떨어짐. 그러나 속도는 좋은 컴퓨터를 사용하면 어느 정도 해결 가능.  
이때 해석력을 높이기 위해 해석 전용 XAI 활용.

### 데이터 준비 단계
##### train / test split
* 데이터를 분석 모형에 넣기 위한 준비과정
* train / test split(공부용 / 시험용 데이터 분리) -> row 분리
* feature / label 분리(문제 부분과 답 부분 분리) -> column 분리
* "A 가 B 에 영향을 줄 것이다" 라는 문장이 있다고 한다면,  
A 에 들어가는 것이 feature, B 가 label
\-> (날짜, 시간, 날씨) 가 이용량에 영향을 줄 것이다.

* 일반적으로 label(답) 은 1개의 컬럼, 나머지는 feature(문제)로 나눈다.  
label 이 늘어날 경우, 리스트를 활용하면 해결 가능.

##### 모델 기획(알고리즘 선택)
* 지도학습은 크게 2가지 시험 유형이 있습니다.
 1. 숫자를 예측
 2. 경우의 수를 예측(개 / 고양이 맞추기 등)
* 숫자를 맞추는 유형은 Regress~ 라는 수식어가 들어갑니다.(Regression / Regressor 등)
* 경우의 수를 맞추는 유형은 Classify~ 가 들어갑니다.(Classification / Classifier 등)
* 숫자를 맞추는 유형을 회귀문제, 경우의 수를 맞추는 유형을 분류 문제라 합니다.
* 이번 데이터 셋의 목적은 숫자(수요량)를 맞추는 것.  
\-> 따라서 회귀 알고리즘을 사용합니다.

Regress~ 의 경우, 이항형 자료를 다루는 Logistic Regression 을 제외하면 모두 숫자를 예측하는 방법.

### 모델 수립
##### 모델을 가져와 생성(학습 이전)
##### 모델 학습(학습용)
##### 모델 평가지표 적용(채점, 학습용과 시험용 모두)

```python
from sklearn.ensemble import RandomForestRegressor as rf
# sklearn -> ML 계열(머신러닝) 알고리즘이 강한 파이썬 라이브러리 -> 대안 少
# 데이터 사이언스 전반으로 많이 쓰임(전처리, 스케일링, ...)

# 알고리즘의 원리와 이론을 정확히 안 후, 머신러닝과 딥러닝의 차이 알아보기
# 일단은 어떤 알고리즘을 어떤 목적에 사용하는지를 이해한 후, 코드 자체도 살펴본다.
# 전통적인 분석 방법과도 비교해보기
model = rf()  # 알고리즘 생성하여 model 변수로 사용
model.fit(X_train, y_train)  # 적합 / 피팅이라고 하며, 학습시킴을 의미.
print(model.score(X_train, y_train), model.score(X_test, y_test))
# 시험 보고, 점수 확인(score = 점수)
# model.score에 기본 적용되는 회귀문제 평가지표는 R-squared 이용.
# 평균으로 예측한 것 대비 얼마나 정답에 더 가까운지를 비율로 나타낸 것.

# 보통은 0 ~ 1 사이, 1에 가까울수록 예측력이 좋다고 판단(실제와 예측이 동일하면 1)
# 0 이면 평균으로 예측한 것과 동일한 정도로 틀렸음(의미 X) 의미.
```

```python
# 가끔 음수가 나올 때가 있는데, 평균으로 답지를 적어 낸 것보다
# 알고리즘 예측값이 더 안 좋다는 의미.

# 오차 제곱합(SSE)이 총 제곱합(SST)보다 클 때,
# 즉, 평균을 예측값으로 활용한 잔차 제곱 합보다 예측 결과가 더 안 좋을 때,
# label 과 features 가 아무런 연관이 없는 경우,
# R-squared 가 음수가 나올 수 있음.

# 전통적인 분석 방법과 알고리즘 분석 방법(머신러닝)의 차이가 아닌,
# 단변량(y가 1개, x는 상관 X) 선형회귀의 임팩트에 따른 오해!
# 통계학에서 더 자세히 복습/증명할 것.
```

### 데이터 준비 단계 조금 더 알아보기 - 변수 선택법
##### Feature selection - 어떤 변수를 알고리즘에 넣을 것인가?
##### Filter / Wrapper / Embed 방식에 대해 다룹니다.
* Filter - 대상 컬럼을 하나 선택하고, 기초통계량으로 판단(ex. 표준편차 0에 근접)
* Wrapper - 컬럼 하나씩 넣기도 하고 빼기도 하면서 평가지표 변화 확인(최적조합 찾기)
* Embed - 알고리즘 중에 내부를 뜯어볼 수 있고, 그 안에 변수의 중요도가 있는 경우  
\--> Tree 계열 또는 Regularizer(패널티 항, L1)가 알고리즘 내부에 있는 경우  
\--> Tree 계열 : DecisionTree, RandomForest, GBDT 류  
\--> 알고리즘 공식문서(document)상 Lasso 또는 L1 이라는 표현이 있는 경우
* 또 다른 방식 : Permutation Importance(최신반에서 다룸) <- XAI에서 사용하는 것과 동일.  
\--> 변수의 중요도에 대해 분석하는 방법이기 때문에  
\--> 블랙박스 상에서 해석력을 높이는 방법으로 사용될 수도 있고,  
\--> 일정 기준을 충족하는 중요한 변수를 선택하는 방법으로도 사용될 수 있음.

기술통계 분야에서는 이외에도 더 많은 접근법이 존재함.

```python
# 전체 feature가 들어있는 상태로 모델 학습 진행 -> 이후 wrapper 적용.
features = ['season', 'holiday', 'workingday',
            'weather', 'temp', 'atemp', 'humidity', 'windspeed',
            'year', 'month', 'day', 'hour', 'dayofweek']
label = 'count'
train, test = df[0::2], df[1::2]
train, test = train.reset_index(), test.reset_index()
X_train, y_train = train[features], train[label]
X_test, y_test = test[features], test[label]
model = rf()
model.fit(X_train, y_train)
```

```python
from sklearn.feature_selection import RFE
# Recursive Feature Elimination -> 하나씩 넣었다 뺐다 하며 점수를 확인하는 방식.
selection = RFE(estimator=model)  # 어떤 알고리즘을 이용해 판단할지 estimator로 결정
selection.fit(X_train, y_train)  # RFE 방식으로 판단(어떤 feature 가 중요할까?)
# 중요한 컬럼이라면 추가 / 제거 시, 점수 변동이 클 것이라 생각하는 것
```

```python
rfe_df = pd.DataFrame()
rfe_df['ranking'] = selection.ranking_  # RFE 방식은 위의 알고리즘들과 달리, 예측값을 출력하지 않고 중요도 순서를 출력함.
rfe_df['features'] = features
rfe_df.sort_values(by='ranking')
# ranking 이 1인 것들은 알고리즘에서 제외시켰을 때 성능 하락이 크다고 판단.
# ranking 이 떨어지는 것들은 알고리즘에서 제외시켰을 때 성능 변화 없거나 상승.

# RFE 방식의 최대 단점: 속도 느림 -> 데이터의 용량이 커지면 다른 방법으로 접근해야 함.
```

```python
type(selection.ranking_)
# 참고로 위에서 selection.ranking_ 속성으로 불러온 결과는 ndarray로
# 데이터프레임으로 변환시 자세한 결과를 알 수 있음.(딕셔너리도 마찬가지)
## 그렇다면 RFE 객체에 어떤 attribute들(예. ranking_)이 있는지 알 수 있는 방법은?
```

feature별로 중요도 파악 가능합니다.

```python
model.feature_importances_ # ndarray 자료형으로 수치가 저장되어 있음
```

```python
# model.feature_importances_ 에는 각 feature 들이 fit(학습) 단계를 거친 후,
# 알고리즘 안에서 얼마나 사용되었는지를 나타내는 수치가 들어있습니다.
# 위의 RFE 와는 상관없이, RandomForest 나 DecidisionTree 등의 Tree 계열 모델을 사용할 때, 자동적으로 수치가 들어있음.
embed_df = pd.DataFrame()
embed_df['importance'] = model.feature_importances_
embed_df['features'] = features
embed_df = embed_df.sort_values(by='importance', ascending=False)
# 만약 wrapper 방식과 embed 방식의 결과가 큰 차이가 있다면 하나의 방식이 잘못된 것.
```
