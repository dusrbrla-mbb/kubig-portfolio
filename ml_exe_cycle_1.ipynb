```python
# 엑셀 파일은 csv 파일에 비해 처리 속도가 느림.
import time
t0 = time.time()
pd.read_csv(path_list[0], encoding='cp949') # csv 파일 열 때
t1 = time.time()
pd.read_excel(path_list[-1]) # excel 파일 열 때
t2 = time.time()

csv_rows = 1000000
excel_rows = 465715
csv_time= t1 - t0
excel_time= t2 - t1
csv_row_per_sec = csv_rows // csv_time
excel_per_sec = excel_rows // excel_time
print(csv_row_per_sec, excel_per_sec)
# 현재 컬럼 개수가 동일하므로, 초당 row 를 처리하는(읽는) 속도로 비교할 수 있다.
# 거의 100 배 차이가 있으므로, 반드시 csv 파일로 변환해야 다음 단계가 빨라진다.
```

```python
# 컬럼 개수와 컬럼명(순서)이 동일한지 확인 -> 하나의 파일로 합칠 수 있는지
# 만약 다르다면 하나로 합치려는 시도 X
all_cols = []
for path in path_list:
    if '.csv' in path:
        col = list(pd.read_csv(path, encoding='cp949').columns)
    else:
        col = list(pd.read_excel(path).columns)
    all_cols = all_cols + col
print(len(all_cols))
# 만약 모든 파일들의 컬럼이 같다면, all_cols 안의 것들을 중복제거시, 총 12개가 남아있어야 함.
```

```python
set(all_cols)
# 파이썬에서의 set 자료형으로 중복 제거.
# 컬럼 이름이 다르게 표기된 것뿐, 실제로는 같은 데이터.
```

```python
all_cols[:12]
# 읽어들였을 때, 컬럼 이름이 다르다 해도, 이 리스트로
# 컬럼 이름을 바꾸면 모든 데이터를 한 데이터 프레임에 넣을 수 있다.
```

```python
reset_col = all_cols[:12]
interested = ['대여일자', '대여시간', '대여구분코드', '성별', '연령대코드']
# 비지니스 이해 단계에서 설정한 목적에 따라 컬럼 선별 -> 추가적으로 용량 감소 가능
all_df = pd.DataFrame()  # 모든 데이터를 담을 빈 데이터 프레임 생성
for path in path_list:
    if '.csv' in path:
        single = pd.read_csv(path, encoding='cp949')
    else:
        single = pd.read_excel(path)
    single.columns = reset_col  # 컬럼 이름 재정의
    all_df = pd.concat([all_df, single[interested]])  # 관심 있는 컬럼만 선별하여 all_df에 row 추가
    print(all_df.shape)  # 컬럼 개수/이름 동일하다고 가정한 후, row 만 추가하는 것(concat)이므로, 컬럼 개수가 늘어나는지 확인.
    del single  # 메모리 관리 코드 -> 대용량 데이터이므로 메모리를 위해, 최종 결과인 all_df 를 제외한 중간 산출물 single 삭제
# 최근에 다루는 데이터는 대용량이므로, 컴퓨터가 획기적으로 바뀌지 않는 이상, 코드 실행 시간과 메모리를 반드시 고려해야 함.
```

```python
all_df.to_csv('processed_17_21.csv', index=False)
# 다 모은 데이터를 한 파일로 저장.
```

### 클렌징한 데이터로 재시작

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dir2 = '/content/drive/My Drive/Study/Jupyter'
df = pd.read_csv(dir2 + '/' + 'clean_data2.csv')
# 파일을 읽고 저장하는 속도는 DISK 의 영향이 가장 큼.
# SSD + M.2 또는 SSD + NVME 선택.
```

```python
df.head()
```

```python
# 이후의 처리 과정은 비지니스 이해 단계에서 정한 목적과 중점적으로 살펴볼 변수 간 관계에 따라 약간 달라짐.
df['count'] = 1
reg = df.loc[df['type'] == 'regular']  # 여기서는 정기권 손님(type 컬럼)들의
df.groupby('datetime').agg({'count' : np.sum})  # 시간대별(datetime 컬럼) 이용 패턴을 보려고 함 -> 목적에 따라 다름.
```

```python
reg = df.loc[df['type'] == 'regular']
reg = reg.groupby('datetime').agg({'count' : np.sum}).reset_index()
reg.columns = ['datetime', 'regular']

cas = df.loc[df['type'] == 'casual']
cas = cas.groupby('datetime').agg({'count' : np.sum}).reset_index()
cas.columns = ['datetime', 'casual']

gro = df.loc[df['type'] == 'group']
gro = gro.groupby('datetime').agg({'count' : np.sum}).reset_index()
gro.columns = ['datetime', 'group']
# 시간대별 이용권별 고객 이용 건수를 각각 3개의 변수로 보관.
```

```python
rc = pd.merge(reg, cas, how='outer')  # 'outer' 는 합집합, 'inner' 는 교집합
rcg = pd.merge(rc, gro, how='outer')
rcg = rcg.fillna(0)  # na(빈칸) 을 채운다고 해서 fillna.
rcg.head()
```

```python
rcg.to_csv('by_time_merge.csv', index=False)
```

### 날씨가 사람의 기분에 영향을 주는 것 반영
##### 불쾌지수(Humidex) 적용
* 온도와 이슬점을 이용하여 땀이 증발하는 정도를 나타냄.
* 땀이 증발하는 속도로 날씨가 기분에 미치는 영향 반영.

```python
T = 40
Td = 20
exp = 2.71828
upper = 5417.753 * ((1 / 273.16) - (1 / (273.15 + Td))) # 공식 대입
exp = 6.11 * (exp ** upper) # 식 간단히
humidex = T + (5 / 9 * (exp - 10)) # 최종 식
humidex
```

```python
def humidex_cal(T, Td):
    exp = 2.71828
    upper = 5417.753 * ((1 / 273.16) - (1 / (273.15 + Td)))
    exp = 6.11 * (exp ** upper)
    humidex = T + (5 / 9 * (exp - 10))
    return humidex
```

```python
df['humidex'] = humidex_cal(df['temp'], df['dew_point'])
```

```python
df['discomfort'] = 0
df.loc[df['humidex'] >= 30, 'discomfort'] = 1
df.loc[df['humidex'] >= 40, 'discomfort'] = 2
df.loc[df['humidex'] >= 46, 'discomfort'] = 3
# 나누는 기준은, 레퍼런스에서 30 -> 40 -> 46 이었기 때문.
df.head()
```

### 날씨가 사람의 건강에 미치는 영향
##### 여름엔 열사병 위험, 겨울엔 동상의 위험
##### heat index(열지수), windchill(겨울 체감온도)
* 열지수가 높으면 열사병 걸릴 확률 증가 -> 외부활동 저하 요인
* 겨울 체감온도는 동상 확률과 연관 -> 자전거 이용시 손을 내놓기 때문.

##### 열지수(Heat Index)

```python
T = 42
R = 40
c1 = -8.78469475556
c2 = 1.61139411
c3 = 2.33854883889
c4 = -0.14611605
c5 = -0.012308094
c6 = -0.0164248277778
c7 = 0.002211732
c8 = 0.00072546
c9 = -0.000003582
TR = T * R
T2 = T * T
R2 = R * R
T2R = T2 * R
TR2 = R2 * T
T2R2 = T2 * R2
part_1 = c1 + (c2 * T) + (c3 * R) + (c4 * TR) + (c5 * T2)
part_2 = (c6 * R2) + (c7 * T2R) + (c8 * TR2) + (c9 * T2R2)
heat_index = part_1 + part_2
heat_index
```

```python
def heat_index_cal(T, R):
    c1 = -8.78469475556
    c2 = 1.61139411
    c3 = 2.33854883889
    c4 = -0.14611605
    c5 = -0.012308094
    c6 = -0.0164248277778
    c7 = 0.002211732
    c8 = 0.00072546
    c9 = -0.000003582
    TR = T * R
    T2 = T * T
    R2 = R * R
    T2R = T2 * R
    TR2 = R2 * T
    T2R2 = T2 * R2
    part_1 = c1 + (c2 * T) + (c3 * R) + (c4 * TR) + (c5 * T2)
    part_2 = (c6 * R2) + (c7 * T2R) + (c8 * TR2) + (c9 * T2R2)
    heat_index = part_1 + part_2
    return heat_index
```

```python
df['heat_index'] = heat_index_cal(df['temp'], df['humidity'])
# 영하의 온도에서는 열사병 위험 X,
# 따라서 열지수는 도표상 27도 이상일 때만 유효.
# 온도가 27도 미만일 경우, 열지수 값을 0으로 보정하여 수식상 오류 수정.
df.head()
```

```python
df.loc[df['temp'] < 27, 'heat_index'] = 20 # df 에 대해 조건을 설정하는 것이므로 df['heat_index'] 대신 사용 가능. 
# 27도 이하일 경우, heat_index 값을 일괄적으로 낮춰서 반영(20 정도)
# 열지수 공식 자체가 27도 이상일 경우에 의미를 갖기 때문에, 정확한 열지수 값은 모르지만,
# 이후에 일괄적으로 열사병 컬럼을 생성할 때, 데이터의 의미가 왜곡되는 것 방지.
df.head()
```

```python
# 27 미만 -> 위험 X
# 27 ~ 32 -> 주의
# 33 ~ 39 -> 매우 주의
# 40 ~ 52 -> 위험
# 53 이상 -> 매우 위험
df['heat_stroke'] = 0
df.loc[df['heat_index'] >= 27, 'heat_stroke'] = 1
df.loc[df['heat_index'] >= 33, 'heat_stroke'] = 2
df.loc[df['heat_index'] >= 40, 'heat_stroke'] = 3
df.loc[df['heat_index'] >= 53, 'heat_stroke'] = 4
# 나누는 기준은, 레퍼런스 참조.
df.head()
```

* Ta 가 온도, v 가 풍속.

```python
# 도메인의 특수성(걷는 것이 아니라 자전거를 타는 것이므로, 체감 풍속은 바뀐다)에 의해,
# 위의 공식은 개조의 여지가 있다.
# 결론적으로, 여기서 하고 싶은 말은,
# 정확하게 측정된 데이터라고 할지라도, 데이터 자체가 변할 수 있는지 신중하게 고려해야 한다.
# 생각을 하는 것과 안 하는 것의 차이가 분명히 발생할 수 있다.
```

```python
T = -20
v = 14
c1 = 13.12
c2 = 0.6215
c3 = -11.37
c4 = 0.3965
c5 = 0.16
vv = v ** c5
windchill = c1 + (c2 * T) + (c3 * vv) + (c4 * T * vv)
windchill
```

```python
def windchill_cal(T, v):
    c1 = 13.12
    c2 = 0.6215
    c3 = -11.37
    c4 = 0.3965
    c5 = 0.16
    v = v + 20  # 자전거라는 도메인을 반영하여, 시속 20km/h 수준으로 달린다 가정.
    vv = v ** c5
    windchill = c1 + (c2 * T) + (c3 * vv) + (c4 * T * vv)
    return windchill
```

```python
df['windchill'] = windchill_cal(df['temp'], df['windspeed'])
df.sort_values(by='windchill')
```

```python
df['frost_bite'] = 0
df.loc[df['windchill'] < -25, 'frost_bite'] = 1
df.loc[df['windchill'] > -25, 'frost_bite'] = 0
df.loc[df['temp'] > 10, 'frost_bite'] = 0
# 겨울 체감온도는 온도가 높을 경우 정의가 되지 않기 때문에,
# 온도가 10도 초과인 경우, frost_bite 에 0 으로 반영.
df.head()
```
